### Ticket promedio por ciudad

```python
# Tipos de datos actualizados
df.dtypes
```

```python
# Conversión de algunas columnas Object a tipo string
df['City']= df['City'].astype('string')
df['Product line']= df['Product line'].astype('string')
df['Gender']= df['Gender'].astype('string')
```

```python
# Mostrar sólo los primeros dos decimales en los cálculos
pd.options.display.float_format= '{:.2f}'.format
```

```python
# Agrupamiento que filtra el total vendido por ciudad
city_sum= df.groupby('City')[['Total']].sum().sort_values(by= 'City', ascending= True)
city_sum
```

```python
city_sum= city_sum.reset_index()
```

```python
# Del agrupamiento anterior se hace la conversión de algunas columnas a arrays, para obtener el ticket promedio con Numpy
city_array= city_sum['City'].to_numpy()
total_array= city_sum['Total'].to_numpy()
```

```python
total_array
```

```python
# Array del conteo de ventas por ciudad, para calcular la media manualmente
city_counts= df['City'].value_counts().sort_index(ascending= True)
city_counts
```

```python
# Cálculo de la media dividiendo el total de compras por ciudad entre los valores de city_counts, ya que .mean() no funciona
for x, y in enumerate(city_counts):
    print(f'{city_array[x]} tiene el ticket promedio de {((total_array[x] / y) * 100):.2f}')
```

```python
# Comparación con el ticket promedio por ciudad, ordenado alfabéticamente en pandas
df.groupby('City')[['Total']].mean().sort_values(by= 'City', ascending= True)
```


### Aporte por categoría (Formas de pago)

```python
# Cantidad de pagos hechos por cada forma
df['Payment'].value_counts()
```

```python
# Agrupación de la sumatoria del total de todas las formas de pago
payment_total= df.groupby('Payment')['Total'].sum()
```

```python
payment_total
```

```python
# Se hace reset al index para ver las columnas en el dataframe
payment_total= payment_total.reset_index()
```

```python
payment_total
```

```python
# Arrays de Numpy de payment y total
payment= payment_total['Payment'].to_numpy()
total= payment_total['Total'].to_numpy()
```

```python
total.sum()
```

```python
# Lista aporte donde se almacena el porcentaje de aporte por categoría
aporte=[]
for i in total:
    aporte.append((i / total.sum()) * 100)
```

```python
# Comprobamos que la suma de los elementos es 100
sum(aporte)
```

```python
payment
```

```python
# Ciclo para imprimir las formas de pago y su porcentaje de aporte redondeado
for index, x in enumerate(aporte):
        print(f'El {round(x)} % de los pagos se han hecho con {payment[index]}.')
```

### Aporte por categoría (Ciudad)

```python
# Cuántas compras se han hecho por ciudad
df['City'].value_counts()
```

```python
# Df de suma de ventas por ciudad, reset en index para utilizar las columnas
city_total= df.groupby('City')['Total'].sum().reset_index()
city_total
```

```python
# Se forman arrays de Numpy del dataframe anterior
city= city_total['City'].to_numpy()
total_city_array= city_total['Total'].to_numpy()
```

```python
# Lista donde se almacena el porcentaje de aporte por ciudad
aporte_city=[]
for i in total_city_array:
    aporte_city.append((i / total_city_array.sum()) * 100)
```

```python
aporte_city
```

```python
# Ciclo para imprimir la ciudad con su respectivo porcentaje de ventas al redondo
for index, x in enumerate(aporte_city):
        print(f'La ciudad {city[index]} aporta con el {round(x)} % de las ventas.')
```

### Matriz de correlación

```python
df.dtypes
```

```python
# Añadimos una columna de hora al dataframe con los primeros dos caracteres de la columna Time (sin minutos)
df['hora']= df['Time'].str[:2]
```

```python
df['hora']
```

```python
# Conversión de hora a tipo entero
df['hora']= df['hora'].astype(int)
```

```python
# Matriz de correlación de algunas columnas
corr= df[['Unit price',	'Quantity',	'Tax 5%', 'Total', 'cogs', 'gross income', 'Rating', 'hora']].corr(numeric_only= True)
```

```python
# Aplicación de background azul
corr.style.background_gradient(cmap= 'Blues')
```

```python
# Matriz de correlación de horas y total
corr_horas= df[['hora', 'Total']].corr()
corr_horas
```

```python
# Heatmap de correlación de horas y total
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr_horas, annot= True, cmap= 'Blues')
plt.title('Heatmap de ventas y horas', fontdict= {'fontsize': 18}, pad= 16)
plt.show()
```

```python
# Matriz de correlación de rating y precio unitario
corr_rating= df[['Rating', 'Unit price']].corr()
```

```python
corr_rating
```

```python
# Heatmap de rating y precio unitario
sns.heatmap(corr_rating, annot= True, cmap= 'Blues')
plt.title('Heatmap de precio unitario y rating', fontdict= {'fontsize': 18}, pad= 16)
plt.show()
```

### Observaciones:
- No existe correlación entre la hora de compra y el total adquirido.<br><br>
- No existe correlación entre precio unitario y rating.<br><br>
- Aunque el precio no influye en el rating, otras variables como el total adquirido, el impuesto de 5 %, el costo de ventas y la ganancia bruta sí están fuertemente correlacionadas entre sí, además de tener correlación alta con la cantidad de compra y el precio unitario: <br><b>Entre más se compra, mayor es el impuesto a pagar.</b>
